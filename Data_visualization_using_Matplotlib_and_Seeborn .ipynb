Lets add more than one line plot in graph.

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years,apples)
plt.plot(years,oranges)

plt.title('Yeilds result')
plt.xlabel('Years')
plt.ylabel('Tons per hectare');
```

Adding legends to understand which line is for which crops.

```python
plt.plot(years,apples)
plt.plot(years,oranges)

plt.xlabel('Years')
plt.ylabel('Tons per hectare')

plt.title('Crop yeild of nation')
plt.legend(['Apples','Oranges']);
```

Styling Lines and Markers
The `plt.plot` function supports many arguments for styling lines and markers:

* `color or c`: Set the color of the line (supported colors)
* `linestyle or ls`: Choose between a solid or dashed line
* `linewidth or lw`: Set the width of a line
* `markersize or ms`: Set the size of markers
* `markeredgecolor or mec`: Set the edge color for markers
* `markeredgewidth or mew`: Set the edge width for markers
* `markerfacecolor or mfc`: Set the fill color for markers
* `alpha`: Opacity of the plot

```python
plt.plot(years,apples, c='b', marker='s', ls='-', lw=2, ms=8, mew=2, mec='navy' )
plt.plot(years,oranges, marker='o', c='r', ls='--',lw=3, alpha=0.5)

plt.xlabel('Years')
plt.ylabel('Tons per hectare')

plt.title('Crop yeild of Pune')
plt.legend(['Apples','Oranges']);
```

We can use `fmt` short form to assign marker,linestyle and color. The syntax of fmt in `plt.plot` is as below
* **`fmt = '[marker][line][color]'`**

```python
plt.plot(years,apples,'s-b')
plt.plot(years,oranges,'o--r')

plt.xlabel('Years')
plt.ylabel('Tons per hectare')

plt.title('Crop yeilds of Pune')
plt.legend(['Apples','Oranges']);
```

# Changing figure size

# Improving default style using seaborn

We can improve styling of plot using `sns.set_style` function from **seaborn**.

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12,6))
plt.plot(years,apples,marker='s',c='b',ls='-',lw=2,ms=5, mew=3,mec='navy')
plt.plot(years,oranges,'o--r')

plt.xlabel('Years')
plt.ylabel('Tons per hectare')

plt.title('Crop yeilds in Pune')
plt.legend(['Apples','Oranges']);
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12,6))
plt.plot(years,apples,marker='s',c='b',ls='-',lw=2,ms=5, mew=3,mec='navy')
plt.plot(years,oranges,'o--r')

plt.xlabel('Years')
plt.ylabel('Tons per hectare')

plt.title('Crop yeilds in Pune')
plt.legend(['Apples','Oranges']);
```

# Scatter plot

In a scatter plot, the values of 2 variable are plotted as a point on 2-dimensional grid. Additionally, you can also use the third variable tod determine size and color of the points.

```python
flower_df = sns.load_dataset('iris')
```

```python
flower_df
```

```python
flower_df.species.unique()
```

If we use `plt.plot` on this data and output will be not as desired. let's try

```python
plt.plot(flower_df.sepal_length,flower_df.sepal_width);
```

As we can see from above grape data is not looking good. Let's try using `scatterplot` from `seaborn` which we imported as `sns`.

```python
sns.scatterplot(flower_df.sepal_length,flower_df.sepal_width);
```

To avoid above note in output we need to use `x=` and `y=` in scatterplot.

```python
sns.scatterplot(x=flower_df.sepal_length, y=flower_df.sepal_width);
```

In above output all points has same color. We can use `hue` option from `scatterplot` to differenciate points on the basis of variable value. Also we can increase size of points using `s=` option.

```python
flower_df.head()
```

```python
sns.scatterplot(x=flower_df.sepal_length,y=flower_df.sepal_width, hue=flower_df['species'],s=100);
```

```python
plt.figure(figsize=(12,6))
plt.title('Sepal Dimensions')

sns.scatterplot(x= flower_df.sepal_length, y= flower_df.sepal_width,hue= flower_df.species, s=100);
```

# Plotting using Pandas DataFrame

`Seaborn` has lot functionality, one of which is we can do plotting using pandas data frame. We need to use `data=` option in scatterplot.

```python
plt.figure(figsize=(12,6))
plt.title('Sepal Dimensions')

sns.scatterplot(x='sepal_length', y='sepal_width', hue='species',s=100, data= flower_df);
```

# Histogram

A histogram represents the distribution of a variable by creating bins (interval) along the range of values and showing vertical bars to indicate the number of observations in each bin.

To generate histogram we use `plt.hist` function. 

```python
plt.title('Sepal widths')
plt.hist(flower_df.sepal_width);
```

# Controling size and number of bins

We can give size of bin or each bins using `bin=` options.

```python
# Specifing the number of bins
plt.hist(flower_df.sepal_width, bins=5);
```

```python
# Specifing the boundaries of each bins
import numpy as np
plt.hist(flower_df.sepal_width,bins=np.arange(2,5,.25));
```

# Creating  multiple histograms

We can create multiple histogram.

```python
flower_df.species.unique()
```

```python
setosa_df = flower_df[flower_df.species == 'setosa']
versicolor_df = flower_df[flower_df.species == 'versicolor']
virginica_df = flower_df[flower_df.species == 'virginica']
```

```python
plt.figure(figsize=(12,6))
plt.hist(setosa_df.sepal_width,alpha=.4, bins=np.arange(2,5,.25))
plt.hist(versicolor_df.sepal_width, alpha=.4, bins=np.arange(2,5,.25));
```

```python
#We can also stack multiple histograms on top of one another.

plt.title('Distribution of sepal width')

plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],
        bins=np.arange(2,5,.25),
        stacked=True )

plt.legend(['Setosa','Versicolor','Virginica']);
```

# Heatmap

A heatmap is used to visualize 2-dimensional data like a matrix or a table using colors. We will use dataset provided by **seaborn** which is `flights` for heatmap example.

```python
flights_df = sns.load_dataset('flights').pivot('month','year','passengers')
flights_df
```

```python
plt.title('Number of passangers(1000s)')
sns.heatmap(flights_df);
```

We can also display the actual values in each block by specifying `annot=True` and using the `cmap` argument to change the color palette.

```python
plt.figure(figsize=(12,6))
plt.title('Number of passangers(1000s)')

sns.heatmap(flights_df,fmt='d',annot=True, cmap='Blues');
```

# Plotting multiple charts in a grid

Matplotlib and Seaborn also support plotting multiple charts in a grid, using `plt.subplots`, which returns a set of axes for plotting.

```python
fig, axes = plt.subplots(2,3, figsize=(12,8))

# Use the axes for plotting
axes[0,0].plot(years,apples,'s-b')
axes[0,0].plot(years,oranges,'o--r')

axes[0,0].set_xlabel('Years')
axes[0,0].set_ylabel('Tones per Hectare')
axes[0,0].set_title('Crop yeilds of Pune')
axes[0,0].legend(['Apples','Oranges']);

#pass the axes into scatter plot
axes[0,1].set_title('Sepal dimension')

sns.scatterplot(x=flower_df.sepal_length, y =flower_df.sepal_width, hue=flower_df.species,
               s=100, ax=axes[0,1]);

# Use the axes for plotting
axes[0,2].set_title('Distribution of Sepal width')

axes[0,2].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width],
              bins=np.arange(2,5,.25),
              stacked = True);

axes[0,2].legend(['Setosa', 'Versicolor', 'Virginica']);

# Pass the axes into seaborn
axes[1,0].set_title('Restaurant bills')
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df, ax=axes[1,0]);

#pass the axes into seaborn
axes[1,1].set_title('Flights data')
sns.heatmap(flights_df,cmap='Blues',ax=axes[1,1]);

# Plot an image using the axes
axes[1,2].set_title('Data Science Meme')
axes[1,2].imshow(img)
axes[1,2].grid(False)
axes[1,2].set_xticks([])
axes[1,2].set_yticks([])

plt.tight_layout(pad=2);
```
